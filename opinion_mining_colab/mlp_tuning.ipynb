# MLP tunning

Ejemplo simple de una red neuronal MLP (modificación de algunos hiper-parámetros)

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

x = np.arange(0.0, 1, 0.01).reshape(-1, 1)
y = np.sin(2 * np.pi * x).ravel()

nn = MLPRegressor(hidden_layer_sizes=(3), 
                  activation='tanh', solver='lbfgs')

n = nn.fit(x, y)
test_x = np.arange(-0.1, 1.1, 0.01).reshape(-1, 1)
test_y = nn.predict(test_x)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, s=5, c='b', marker="o", label='real')
ax1.plot(test_x,test_y, c='r', label='NN Prediction')

plt.legend()
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

x = np.arange(0.0, 1, 0.01).reshape(-1, 1)
y = np.sin(2 * np.pi * x).ravel()

nn = MLPRegressor(hidden_layer_sizes=(3), 
                  activation='relu', 
                  solver='sgd', 
                  learning_rate='adaptive', 
                  learning_rate_init=0.001
                 )

n = nn.fit(x, y)
test_x = np.arange(-0.1, 1.1, 0.01).reshape(-1, 1)
test_y = nn.predict(test_x)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, s=5, c='b', marker="o", label='real')
ax1.plot(test_x,test_y, c='r', label='NN Prediction')

plt.legend()
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

def cubic(x):
    return x**3 + x**2 - x - 1


x = np.arange(-2, 2, 0.1).reshape(-1, 1)
y = cubic(x).ravel()

nn = MLPRegressor(hidden_layer_sizes=(6),activation='tanh', solver='lbfgs')

n = nn.fit(x, y)
test_x = np.arange(-2.5, 2.5, 0.01).reshape(-1, 1)
test_y = nn.predict(test_x)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, s=5, c='b', marker="o", label='real')
ax1.plot(test_x,test_y, c='r', label='NN Prediction')

plt.legend()
plt.show()
```
